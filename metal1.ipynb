```python
import pyEPR as epr
from qiskit_metal import MetalGUI, Dict, Headings
from qiskit_metal import designs, draw
import qiskit_metal as metal
%load_ext autoreload
%autoreload 2
```

```python
design = designs.DesignPlanar({}, True)
design.chips.main.size['size_x'] = '4mm'
design.chips.main.size['size_y'] = '6mm'

gui = MetalGUI(design)
```

```python
from qiskit_metal.qlibrary.qubits.transmon_pocket import TransmonPocket
from qiskit_metal.qlibrary.tlines.meandered import RouteMeander
```

```python
TransmonPocket.get_template_options(design)
```

```python

RouteMeander.get_template_options(design)
```

```python
q1 = TransmonPocket(design, 'Q1', options=dict(
    pad_width='425 um',
    pocket_height='650um',
    connection_pads=dict(
        readout=dict(loc_W=+1, loc_H=+1, pad_width='200um')
    )))

q2 = TransmonPocket(design, 'Q2', options=dict(
    pos_x='2.0 mm',
    pad_width='425 um',
    pocket_height='650um',
    connection_pads=dict(
        readout=dict(loc_W=-1, loc_H=+1, pad_width='200um')
    )))

q3 = TransmonPocket(design, 'Q3', options=dict(
    pos_x='2.0 mm',
    pos_y='2.0 mm',
    pad_width='425 um',
    pocket_height='650um',
    connection_pads=dict(
        readout=dict(loc_W=-1, loc_H=-1, pad_width='200um')
    )))

q4 = TransmonPocket(design, 'Q4', options=dict(
    pos_y='2.0 mm',
    pad_width='425 um',
    pocket_height='650um',
    connection_pads=dict(
        readout=dict(loc_W=+1, loc_H=-1, pad_width='200um')
    )))
```

```python
coupler_options1 = Dict(
    pin_inputs=Dict(
        start_pin=Dict(component='Q1', pin='readout'),
        end_pin=Dict(component='Q2', pin='readout')),
    fillet='99.9um',
    total_length='5mm',
    lead=Dict(start_straight='200um'
              ))

coupler_options2 = Dict(
    pin_inputs=Dict(
        start_pin=Dict(component='Q3', pin='readout'),
        end_pin=Dict(component='Q4', pin='readout')),
    fillet='99.9um',
    total_length='5mm',
    lead=Dict(start_straight='200um'
              ))

coupler_options3 = Dict(
    pin_inputs=Dict(
        start_pin=Dict(component='Q1', pin='readout'),
        end_pin=Dict(component='Q3', pin='readout')),
    fillet='99.9um',
    total_length='5mm',
    lead=Dict(start_straight='200um'
              ))

coupler_options4 = Dict(
    pin_inputs=Dict(
        start_pin=Dict(component='Q2', pin='readout'),
        end_pin=Dict(component='Q4', pin='readout')),
    fillet='99.9um',
    total_length='5mm',
    lead=Dict(start_straight='200um'
              ))

bus0 = RouteMeander(design, 'coupler1', options=coupler_options1)
bus1 = RouteMeander(design, 'coupler2', options=coupler_options2)
bus2 = RouteMeander(design, 'coupler3', options=coupler_options3)
bus3 = RouteMeander(design, 'coupler4', options=coupler_options4)

gui.rebuild()
gui.autoscale()
```
